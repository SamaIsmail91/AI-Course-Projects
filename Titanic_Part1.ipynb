### KNN Imputer

```python
# !pip install sklearn
```

```python
from sklearn.impute import KNNImputer
```

```python
# Select relevant features, ensure there are no missing values in other columns that will be used by KNN
df = titanic_df[['Age', 'Pclass', 'SibSp', 'Parch', 'Fare', 'sex_label']]
df.dropna(subset=['Pclass', 'SibSp', 'Parch', 'Fare', 'sex_label'], inplace=True)
```

```python
# Initialize the KNN Imputer
knn_imputer = KNNImputer(n_neighbors=5)

# Perform the imputation; note that KNN imputer returns a numpy array
imputed_data = knn_imputer.fit_transform(df)

# Convert the numpy array back to a pandas DataFrame
imputed_df = pd.DataFrame(imputed_data, columns=df.columns)

imputed_df.head()
```
